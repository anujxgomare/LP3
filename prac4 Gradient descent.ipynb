```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x + 3)**2

def df(x):
    return 2 * (x + 3)  # derivative of (x + 3)^2
```

```python
x = 2               # starting point
learning_rate = 0.1 # step size
epochs = 25         # number of iterations
```

```python
x_values = [x]
for i in range(epochs):
    grad = df(x)                   # find derivative
    x = x - learning_rate * grad   # update x
    x_values.append(x)
    print(f"Iteration {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")
```

```python
plt.plot(x_values, [f(i) for i in x_values], 'o-')
plt.title('Gradient Descent: Convergence to Local Minima')
plt.xlabel('x value')
plt.ylabel('f(x)')
plt.show()
```

```python
print("\nApproximate local minima occurs at x =", round(x, 4))
```
